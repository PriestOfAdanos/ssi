```python
class Fuzzy:
    def __init__(self):
        self.antecedent=[]
        self.consequent=[]
        self.rules=[]
        
    def addAntecedent(self, parametr, linquisticValue, a, b, c):
        self.antecedent.append([parametr, linquisticValue, a, b, c])
        
    def addConsequent(self, linquisticValue, param): #['jakosc zycia',[['poor',a,b,c],['avg',a1,b1,c1],['good',a2,b2,c2]]]
        self.consequent.append([linquisticValue, param])   
        
    def addRule(self,rule):
        self.rules.append(rule)
        
    def triangularFunction(self,x,a,m,b):
        if x<=a:
            return 0
        if x>a and x<=m:
            return ((x-a)/(m-a))
        if x>m and x<=b:
            return ((b-x)/(b-m))
        if x>b:
            return 0
        
    def compute(self, sample):
        
        #rozmycie
        #liczenie funkcji dla poor, avg, good
        
        #tablica obliczonych elementów
        fuzzyValues=[]
        for i in self.antecedent:
            fuzzyValues.append([i[0],i[1], self.triangularFunction(sample[i[0]],i[2],i[3],i[4])])
        
        #wnioskowanie
        
        #korzystanie z reguł
        rulesResult=[]
        for i in self.rules:
            tmp = 1
            for j in fuzzyValues:
                if i[j[0]] == j[1]:
                    tmp *= j[2]
            rulesResult.append(tmp)
        
        #wyostrzenie
        
        #metoda środka zbioru rozmytego
        # http://dydaktyka.polsl.pl/kwmimkm/InzW_W_WnioskowanieRozmyte.pdf strona 39
        # suma(wartośćReguły * maximum(wartościReguł))/suma(wartosciRegul)
        
        print(rulesResult)
        tmp = []
        maxRules = max(rulesResult)
        for i in rulesResult:
            tmp.append(i*maxRules)
        result = sum(tmp)/sum(rulesResult)
        
        #odczytanie wartości z wykresu oraz zwrócenie/wyświetlenie wyniku
        
        #zwrócenie max względem następnika
        decision = ''
        tmp = 0
        for i in self.consequent[0][1]:
            if self.triangularFunction(result,i[1],i[2],i[3]) > tmp:
                tmp = self.triangularFunction(result,i[1],i[2],i[3])
                decision = i[0]
                
        print("Jakość życia to",decision)
        
```

```python
class ProcessingData:
    @staticmethod
    def stiffle(x):
        for i in range(len(x)-1,0,-1):
            tmp = random.randint(0,i)
            x.iloc[i], x.iloc[tmp] = x.iloc[tmp], x.iloc[i]
        return x
    
    @staticmethod
    def normalize(X):
        x=X.copy()
        values = x.select_dtypes(exclude="object")
        indexes = values.columns.tolist()
        
        for i in indexes:
            xmax = max(x[i])
            xmin = min(x[i])
            mnx = xmax - xmin
            for v in range(len(x[i])):
                x.at[v,i] = (x.at[v,i]-xmin)/mnx
        return x
        
    @staticmethod
    def split(x,a):
        n = len(x)
        tabA=x[0:int(a*n)]
        tabB=x[int(a*n):n]
        return tabA, tabB
```

```python
class irisFuzzy:
    def __init__(self):
        self.antecedent=[]
        self.consequent=[]
        self.rules=[]
        
    def addAntecedent(self, parametr, linquisticValue, a, b, c, d):
        self.antecedent.append([parametr, linquisticValue, a, b, c, d])
        
    def addConsequent(self, linquisticValue, param): #['variety',[['Setosa',a,b,c,d],['Versicolor',a1,b1,c1,d1],['Virginica',a2,b2,c2,d2]]]
        self.consequent.append([linquisticValue, param])   
        
    def addRule(self,rule):
        self.rules.append(rule)
        
    def trapezeFunction(self,x,a,m,n,b):
        if x<=a:
            return 0
        if x>a and x<=m:
            return ((x-a)/(m-a))
        if x>m and x<=n:
            return 1
        if x>n and x<=b:
            return ((b-x)/(b-n))
        if x>b:
            return 0
        
    def compute(self, sample):
        #rozmycie
        
        # sample = {'sepal.length','sepal.width','petal.length','petal.width'}
        
        #tablica obliczonych elementów
        fuzzyValues=[]
        for i in self.antecedent:
            fuzzyValues.append([i[0],i[1], self.trapezeFunction(sample[i[0]],i[2],i[3],i[4],i[5])])
        
        #wnioskowanie
        
        #korzystanie z reguł
        x=[] 
        for i in sample:
            x.append(sample[i])

        rulesResult=[]
        
        #wybranie wag = min()
        for i in self.rules:
            tmp = 2
            for j in fuzzyValues:
                if j[0] in i:
                    if i[j[0]] == j[1]:
                        if tmp > j[2]:
                            tmp = j[2]
            y = eval(i['Variety'])
            rulesResult.append([tmp,y])       

        #wyostrzenie
        vulgar = 0
        denominator = 0
        
        #wyliczenie średniej ważonej
        for i in rulesResult:
            vulgar += i[0]*i[1]
            denominator += i[0]
        result = vulgar/denominator
        
        #zabezpieczenie przez przypadkowym NaN
        if np.isnan(result):
            result = 0
            
        #odczytanie wartości z wykresu oraz zwrócenie/wyświetlenie wyniku
        
        #zwrócenie max względem następnika
        decision = ''
        tmp = 0
        for i in self.consequent[0][1]:
            if self.trapezeFunction(result,i[1],i[2],i[3],i[4]) > tmp:
                tmp = self.trapezeFunction(result,i[1],i[2],i[3],i[4])
                decision = i[0]
                
        return decision
        
```

```python
#losowanie irysów

IrisMixed=ProcessingData.normalize(ProcessingData.stiffle(iris))
irisTrain, irisVal = ProcessingData.split(IrisMixed,0.7)

#obróbka danych treningowych

systemI=irisFuzzy()
irisVarieties = ['Setosa','Versicolor','Virginica']
irisProperties = ['sepal.length','sepal.width','petal.length','petal.width']
irises = []
for i in irisVarieties:
    irises.append([])
    ind = len(irises)-1
    data = np.where(irisTrain.variety == i)[0]
    for x in irisProperties:
        irises[ind].append([])
        for v in data:
            irises[ind][-1].append(irisTrain.iloc[v][x])

consequences = []

#tworzenie poprzedników i następnika

for v in irises:
    sumMin = 0
    sumMax = 0
    sumMean = 0
    sumMedian = 0
    for i in range(len(irisProperties)):
        if (irises.index(v)==0 and i!=1) or (irises.index(v)==1 and i==1):
            minV = 0
            meanV = 0
        elif irises.index(v)!=0 and i==3:
            minV = min(v[0:len(v)][i]) - 0.1
            meanV = np.mean(v[0:len(v)][i])
        else:
            minV = min(v[0:len(v)][i])
            meanV = np.mean(v[0:len(v)][i])
            
        if (irises.index(v)==0 and i==1) or (irises.index(v)==2 and i!=1):
            medianV = 1
            maxV = 1
        elif irises.index(v)!=2 and i==3:
            medianV = np.median(v[0:len(v)][i]) +0.1
            maxV = max(v[0:len(v)][i])
        else:
            medianV = np.median(v[0:len(v)][i])
            maxV = max(v[0:len(v)][i])

        if(meanV>medianV):
            systemI.addAntecedent(irisProperties[i],irisVarieties[irises.index(v)],minV,medianV,meanV,maxV)
        else:
            systemI.addAntecedent(irisProperties[i],irisVarieties[irises.index(v)],minV,meanV,medianV,maxV)
        
        if irises.index(v)==0:
            sumMin = 0
            sumMean = 0
        else:
            sumMin += minV/len(irisProperties)
            sumMean += meanV/len(irisProperties)
        if irises.index(v)==2:
            sumMax = 1
            sumMedian = 1
        else:
            sumMax += maxV/len(irisProperties)
            sumMedian += medianV/len(irisProperties)
        
    if sumMean>sumMedian:
        consequences.append([irisVarieties[irises.index(v)],sumMin,sumMedian,sumMean,sumMax])
    else:
        consequences.append([irisVarieties[irises.index(v)],sumMin,sumMean,sumMedian,sumMax])

systemI.addConsequent('Variety',consequences)

#reguły w systemie Takagi-Sugeno
#rule = {sL/sW/pL/pW : 'variety', ... , 'Variety':'wzór'}
#stworzone na podstawie wykresów z laboratoriów 1 i metody prób i błędów (współczynniki we wzorach)

systemI.addRule({'petal.width':'Setosa','Variety':'2*x[3]'}) 
systemI.addRule({'petal.length':'Setosa','Variety':'2*x[2]'})
systemI.addRule({'sepal.length':'Setosa','petal.length':'Setosa','Variety':'1.5*(x[0]+x[2])'})

systemI.addRule({'petal.width':'Versicolor','petal.length':'Versicolor','Variety':'0.5*(x[2]+x[3])'})
systemI.addRule({'petal.width':'Virginica','petal.length':'Virginica','Variety':'0.5*(x[2]+x[3])'})

systemI.addRule({'petal.width':'Versicolor','petal.length':'Virginica','Variety':'0.5*(x[2]+x[3])'})
systemI.addRule({'petal.width':'Virginica','petal.length':'Versicolor','Variety':'0.125*(x[2]+x[3])'})

systemI.addRule({'sepal.length':'Setosa','petal.width':'Virginica','petal.length':'Virginica','Variety':'1.0/2.0*(x[1]+x[2]+x[3])'})
systemI.addRule({'sepal.length':'Setosa','petal.width':'Versicolor','petal.length':'Versicolor','Variety':'1.0/2.0*(x[1]+x[2]+x[3])'})

#badanie skuteczności

accuracy = 0
setA = 0
virA = 0
verA = 0
for i in irisVal.iloc:
    sample = {}
    for x in irisProperties:
        sample.update({x:i[x]})
    if(systemI.compute(sample)==i.variety):
        accuracy+=1
        if i.variety == 'Setosa':
            setA+=1
        elif i.variety == 'Versicolor':
            verA+=1
        else:
            virA+=1

print('accuracy:',accuracy/len(irisVal)*100,'%')
print('accuracy Setosa:',setA/len(np.where(irisVal.variety == 'Setosa')[0])*100,'%')
print('accuracy Versicolor:',verA/len(np.where(irisVal.variety == 'Versicolor')[0])*100,'%')
print('accuracy Virginica:',virA/len(np.where(irisVal.variety == 'Virginica')[0])*100,'%')
```
